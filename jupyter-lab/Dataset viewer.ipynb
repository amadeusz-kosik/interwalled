```python
import os

import pandas as pd
import matplotlib.pyplot as plt

MAX_ROWS = 100

pd.set_option('display.max_rows', MAX_ROWS)
```

```python
def peek_dataset(path: str): 
    display(pd.read_parquet(path).query('key == "CH-1"').sort_values(by=['from', 'to', 'value']).head(MAX_ROWS))

def peek_result(path: str): 
    df = pd.read_parquet(path)
    df = df.query('key == "CH-1"')
    df = pd.concat([
        df['key'], 
        pd.DataFrame(list(df['lhs'])).drop(['key'], axis=1).add_prefix('lhs_'), 
        pd.DataFrame(list(df['rhs'])).drop(['key'], axis=1).add_prefix('rhs_')
    ], axis=1)
    df = df.sort_values(by=['key', 'lhs_from', 'rhs_from'])
    display(df.head(MAX_ROWS))
```

```python
peek_dataset("../data/sparse-16/100/16/database.parquet/")
```
