```python
import os

import datetime
import pandas as pd
import pathlib
```

```python
pd.options.display.float_format = '{:20,.2f}'.format

def load_data(df_path: str):
    parquet_files = sorted(pathlib.Path(df_path).glob("*.parquet"))
    df_data = pd.concat([pd.read_parquet(p) for p in parquet_files])
    df_data['width'] = df_data['pos_end'] - df_data['pos_start']

    return df_data
    

def generate_stats(df, column_name: str):
    return {
        'mean'          : df[column_name].mean(),
        'median'        : df[column_name].median(),
        'std'           : df[column_name].std(),
        'min'           : df[column_name].min(),
        'max'           : df[column_name].max(),
        '25th quantile' : df[column_name].quantile(0.25),
        '50th quantile' : df[column_name].quantile(0.50), 
        '75th quantile' : df[column_name].quantile(0.75),
        '90th quantile' : df[column_name].quantile(0.90),
        '95th quantile' : df[column_name].quantile(0.95),
    }
```
